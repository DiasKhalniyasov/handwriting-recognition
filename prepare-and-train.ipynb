```python
import os
import json
import sys
import warnings
from NomeroffNet.Base import OCR

warnings.filterwarnings('ignore')

DATASET_PATH = 'C:\\Users\\diask\\cropped_words'
#where files are stored

DATASET_NAME = "handwriting_recognition"
VERSION = "1"
MODE = "gpu"
RESULT_MODEL_PATH = os.path.join(os.path.abspath('../'), "diask", 'hand_ocr_{}_{}-{}.h5'.format(DATASET_NAME, VERSION, MODE))
#where to save trained model
```

```python
target_list = ["Казахстан","Белоруссия","Киргизия","Таджикистан",
               "Узбекистан","Астана","Алматы","Актау","Актобе",
               "Атырау","Караганда","Костанай","Кызылорда","Павлодар",
               "Петропавловск","Семей","Тараз","Уральск","Туркестан",
               "Кокшетау","Шымкент","Россия","Украина","Абхазия",
               "Южная Осетия","Молдавия","Акмолинская","Алматинская",
               "Мангистауская","Актюбинская","Атырауская",
               "Карагандинская","Костанайская","Кызылординская",
               "Павлодарская","Северно-Казахстанская","Восточно-Казахстанская",
               "Жамбыльская","Западно-Казахстанская","Туркестанская",
               "Усть-Каменогорск","Южно-Казахстанская"]
```

```python
#create json file to each image
def save_json (name,start):     
    filepath = 'C:\\Users\\diask\\cropped_words\\val\\img'
    for filename in os.listdir(filepath):
        if filename.endswith(".png"):
            if filename.startswith(start):
                filename = os.path.splitext(filename)[0]
                my_details = {
                  "description": name,
                  "name":   filename
                } 
                #in each json file there is a dict with the name and what is written

                with open('C:\\Users\\diask\\cropped_words\\val\\ann\\%s.json' % filename, 'w') as a:
                    json1 = json.dumps(my_details, ensure_ascii=False)
                    a.write(json1)
```

```python
for i in range(len(target_list)):
    save_json(target_list[i],"{}_".format(i))
```

```python
class kz(OCR): #our model
    def __init__(self):
        OCR.__init__(self)
        
        self.EPOCHS = 3
        # number of epochs to train
```

```python
#setting up the model
ocrTextDetector = kz() 
```

```python
#preparing the data (checking if letters match in train/val/test, creating dictionary of letters, preparing images )
model = ocrTextDetector.prepare(DATASET_PATH, aug_count=0)
```

```python
#training the model
model = ocrTextDetector.train(mode=MODE)
```

```python
#identifying the accuracy on test
ocrTextDetector.test(verbose=True)
```

```python
#saving trained model
ocrTextDetector.save(RESULT_MODEL_PATH, verbose=True)
```
